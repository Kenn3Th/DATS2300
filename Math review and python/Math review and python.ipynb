```python
def signal(omega, time):
    return 2*np.sin(omega*time+1)
```

# Least square polinomial fitting

$y = 0.4x^3 + x^2 - 1.5x + 1 + w$

```python
def signals(x, w):
    return 0.4*x**3 + x**2 - 1.5*x + 1 + w
```

```python
N = 50
sigma = 0.5
mu = 0
x = np.linspace(-1,1,N)
noise = sigma*np.random.randn(N)
y = signals(x,noise)
```

```python
fig = plt.figure()

ax = fig.add_subplot(1,1,1)
ax.plot(x,signals(x,noise),'ro-')
```

$A = \begin{bmatrix}
x^3_1 & x^2_1 & x_1 & 1\\
x^3_2 & x^2_2 & x_2 & 1\\
\vdots&\vdots&\vdots&\vdots\\
x^3_N & x^2_N & x_N & 1\\
\end{bmatrix}$

```python
A = np.zeros((N,4))
for i in range(N):
    A[i,0] = x[i]**3
    A[i,1] = x[i]**2
    A[i,2] = x[i]
    A[i,3] = 1
```

```python
ATAinv = np.linalg.inv(A.T.dot(A))
theta = ATAinv.dot(A.T).dot(y)
theta.shape = (4,1)
theta
```

```python
estimate = A.dot(theta)
```

```python
fig = plt.figure()

axSignal = fig.add_subplot(1,1,1)
axSignal.plot(x,signals(x,noise),'ro-')
axSignal.plot(x,estimate, 'b-')
```

$B = \begin{bmatrix}
x^2_1 & x_1 & 1\\
x^2_2 & x_2 & 1\\
\vdots & \vdots & \vdots\\
X^2_N & X_N & 1
\end{bmatrix}$

```python
B = np.zeros((N,3))
for i in range(N):
    B[i,0] = x[i]**2
    B[i,1] = x[i]
    B[i,2] = 1
```

```python
BTBinv = np.linalg.inv(B.T.dot(B))
theta2 = BTBinv.dot(B.T).dot(y)
theta2.shape = (3,1)
estimate2 = B.dot(theta2)
```

```python
fig = plt.figure()

ax_pol3 = fig.add_subplot(1,1,1)
ax_pol3.set_title('Comparing 3. and 2. order interpolation')
ax_pol3.plot(x,signals(x,noise),'ro-', label = 'Signal')
ax_pol3.plot(x,estimate, 'b-', label = '3. order interpolation')
ax_pol3.plot(x,estimate2, 'y-', label = '2. order interpolation')
ax_pol3.legend()
```

$A = \begin{bmatrix}
x^4_1 & x^3_1 & x^2_1 & x_1 & 1\\
x^4_1 & x^3_2 & x^2_2 & x_2 & 1\\
\vdots&\vdots&\vdots&\vdots&\vdots\\
x^4_N & x^3_N & x^2_N & x_N & 1\\
\end{bmatrix}$

```python
C = np.zeros((N,5))
for i in range(N):
    C[i,0] = x[i]**4
    C[i,1] = x[i]**3
    C[i,2] = x[i]**2
    C[i,3] = x[i]
    C[i,4] = 1
```

```python
CTCinv = np.linalg.inv(C.T.dot(C))
theta3 = CTCinv.dot(C.T).dot(y)
theta3.shape = (5,1)
estimate3 = C.dot(theta3)
```

```python
fig = plt.figure()

ax_pol4 = fig.add_subplot(1,1,1)
ax_pol4.set_title('Comparing 4., 3. and 2. order interpolation')
ax_pol4.plot(x,signals(x,noise),'ro-', label = 'Signal')
ax_pol4.plot(x,estimate, 'k--', label = '4. order interpolation')
ax_pol4.plot(x,estimate, 'y-', label = '3. order interpolation')
ax_pol4.plot(x,estimate2, 'b-.', label = '2. order interpolation')
ax_pol4.legend()
```

# Maps and functions

```python
def R3toUnitSphere(x):
    denom = np.sqrt(x.T.dot(x))
    return x/denom
```

```python
N = 100
sigma = 2
x = sigma*np.random.randn(N,3)

y = np.zeros((N,3))

for i in range(N):
    y[i,:] = R3toUnitSphere(x[i,:])
```

```python
from mpl_toolkits import mplot3d
#Sets figure size
width = 10
height = 8
plt.rcParams['figure.figsize'] = [width, height] 
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter3D(x[:,0],x[:,1],x[:,2])
ax.scatter3D(y[:,0],y[:,1],y[:,2])
```
